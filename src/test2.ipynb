```python
from dataset import GroupImageDataset
import torch
from matplotlib import pyplot as plt
from random import randint
from torchvision import transforms

transform = transforms.Compose([
    transforms.Resize((800, 800)),  # Resize all images to 224x224
    transforms.ToTensor()           # Convert the image to a PyTorch tensor
])

DATASET_DIR = '../dataset/'
sample_test_dataset = GroupImageDataset(DATASET_DIR + 'sample_test.csv', DATASET_DIR + 'sample_test_images/', transform=transform, is_test=True)
```

```python
# sample = sample_test_dataset[randint(0, len(sample_test_dataset))]
sample = sample_test_dataset[84]
image = sample['image']
group_id = sample['group_id']
entity_name = sample['entity_name']
# entity_value = sample['entity_value']

# Plot the image using matplotlib
plt.imshow(image.permute(1, 2, 0))
print(image.shape)
plt.title(f"Group ID: {group_id}\n{entity_name}:")
plt.axis('off')  # Hide axis
plt.show()
```
